## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Making a copy of the dataset
df_clean = df.copy()

# 1. Remove Duplicate Rows

duplicate_count = df_clean.duplicated().sum()
df_clean.drop_duplicates(inplace=True)

print(f"Removed {duplicate_count} duplicate rows.")

# 2. Handling Missing Values

# Numerical columns → fill missing with median
num_cols = df_clean.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    df_clean[col].fillna(df_clean[col].median(), inplace=True)

# Categorical columns → fill missing with mode
cat_cols = df_clean.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)

print("Missing values handled successfully.")

# 3. Converting Date Columns (if present)

date_cols = ['issue_reported_time', 'issue_resolved_time']
for col in date_cols:
    if col in df_clean.columns:
        df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')

# 4. Feature Engineering – Creating Resolution Time
if 'issue_reported_time' in df_clean.columns and 'issue_resolved_time' in df_clean.columns:
    df_clean['resolution_time'] = (
        (df_clean['issue_resolved_time'] - df_clean['issue_reported_time'])
        .dt.total_seconds() / 3600
    )

# Handle negative or missing resolution times
if 'resolution_time' in df_clean.columns:
    df_clean['resolution_time'] = df_clean['resolution_time'].apply(
        lambda x: abs(x) if pd.notnull(x) else x
    )
    df_clean['resolution_time'].fillna(df_clean['resolution_time'].median(), inplace=True)

# 5. Fixing invalid values for handling_time
if 'handling_time' in df_clean.columns:
    df_clean['handling_time'] = df_clean['handling_time'].apply(lambda x: abs(x))
    df_clean['handling_time'].replace(0, df_clean['handling_time'].median(), inplace=True)

print("Data wrangling completed. Dataset is now analysis-ready!")
````

### What all manipulations have you done and insights you found?

To make the dataset analysis-ready, several cleaning and transformation steps were performed. These steps ensure that the data is consistent, accurate, and suitable for Exploratory Data Analysis (EDA) and Machine Learning (ML).

## 1. Removed Duplicate Records
Duplicate rows were identified and removed to avoid repetition of the same ticket.  
**Insight:**  
A few duplicates existed, which could have created biased insights and incorrect model learning.

## 2. Handled Missing Values
- **Numerical columns** were imputed using the **median**, preventing skewness caused by extreme values.  
- **Categorical columns** were imputed using the **mode**, ensuring that the most common category remains consistent.

**Insight:**  
Multiple columns had missing entries, which indicates incomplete data capturing in the customer support workflow.

## 3. Converted Date-Time Columns
`issue_reported_time` and `issue_resolved_time` were converted to datetime format for proper time-based calculations.

**Insight:**  
Some issues had incorrect or missing timestamps, affecting the accuracy of resolution time measurement.

## 4. Created a New Feature: Resolution Time
A new feature **resolution_time (in hours)** was created by calculating the difference between reported and resolved timestamps.

**Insight:**  
Resolution time varied heavily across tickets. Longer resolution duration was found to be associated with lower customer satisfaction (CSAT).

## 5. Cleaned Invalid Handling Time Values
- Negative handling times were corrected to positive values.  
- Zero handling times (not possible in real support workflows) were replaced with median values.

**Insight:**  
Data entry errors existed in the operational logs, showing inconsistency in recording handling time.
